```python
import cv2

image = cv2.imread("./Orginals/Peace-art.jpg")
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
# show the original and blurred images
cv2.imshow("Blurred", blurred)

wide = cv2.Canny(blurred, 10, 200)
mid = cv2.Canny(blurred, 30, 150)
tight = cv2.Canny(blurred, 240, 250)
# show the output Canny edge maps
cv2.imshow("GaussianBlur | Wide Edge Map", wide)
cv2.imshow("GaussianBlur | Mid Edge Map", mid)
cv2.imshow("GaussianBlur | Tight Edge Map", tight)

wide = cv2.Canny(gray, 10, 200)
mid = cv2.Canny(gray, 30, 150)
tight = cv2.Canny(gray, 240, 250)
# show the output Canny edge maps
cv2.imshow("Wide Edge Map", wide)
cv2.imshow("Mid Edge Map", mid)
cv2.imshow("Tight Edge Map", tight)
cv2.waitKey(0)
```
