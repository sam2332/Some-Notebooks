```python
import cloudscraper

requests = cloudscraper.create_scraper(browser="chrome", interpreter="nodejs")
```

```python
requests.post(
    "https://www.gasbuddy.com/graphql",
    {
        "operationName": "LocationBySearchTerm",
        "variables": {"fuel": 1, "maxAge": 0, "search": "48864"},
        "query": "query LocationBySearchTerm($search: String) { locationBySearchTerm(search: $search) { trends { areaName country today todayLow } } }",
    },
).json()
```
