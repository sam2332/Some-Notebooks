```python
class Base:
    def __init__(self, base, prepends=None, appends=None):
        self.base = base
        if prepends is None:
            prepends = [""]
        self.prepends = prepends
        if appends is None:
            appends = []
        self.appends = appends

    def toList(self):
        out = []
        for pp in self.prepends:
            for ap in self.appends:
                out.append(f"{pp}{self.base}{ap}")
            if pp != "":
                out.append(f"{pp}{self.base}")
        out.append(self.base)

        return out
```

```python
numbers = [10, 15, 20, 22, 25, 30, 35]
words = [
    "save",
    "FALL",
    "SUMMER",
    "SPRING",
    "welcome",
    "WINTER",
    "SORRY",
    "MILITARY",
    "Turkey",
    "Love",
    "Santa",
    "SPONSOR",
]
start = [""]
out = []
for w in words:
    out.extend(Base(w, prepends=start, appends=numbers).toList())
out.reverse()
out
```
