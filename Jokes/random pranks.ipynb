```python
from playsound import playsound
import requests
import random
import time
from pynput.keyboard import Key,Controller
from pynput.keyboard import Controller as KeyboardController
from pynput.mouse import Controller as MouseController

from gtts import gTTS
import os

import pywinctl
keyboard = KeyboardController()
mouse = MouseController()
```

```python
class InvertedMouseMover():
    def __init__(self, mouse):
        self.x = mouse.position[0]
        self.y = mouse.position[1]
        self.mouse = mouse

    def getPos(self):
        return (self.x, self.y)

    def update(self):
        self.x += (self.x-self.mouse.position[0])
        self.y += (self.y-self.mouse.position[1])


m = InvertedMouseMover(mouse)
```

```python
def getActiveWindowAppName():
    return pywinctl.getActiveWindow().getAppName()


def getActiveWindowTitle():
    return pywinctl.getActiveWindowTitle()
```

```python
while True:
    if random.choice(chance_volume):
        print('adjusting vol:', end='')
        times =random.randint(1, 5)
        if random.choice(['-', '+']) == "+":
            print(f"up x {times}")
            for _ in range(1, random.randint(1,5)):
                keyboard.press(Key.media_volume_up)
                keyboard.release(Key.media_volume_up)
                time.sleep(0.3)
        else:
            print(f'down x {times}')
            for _ in range(1, random.randint(1,5)):
                keyboard.press(Key.media_volume_down)
                keyboard.release(Key.media_volume_down)
                time.sleep(0.3)

    if random.choice(chance_desktop):
        print("sending to desktop")
        keyboard.press(Key.cmd)
        keyboard.press(Key.d)
        time.sleep(0.25)
        keyboard.release(Key.d)
        keyboard.release(Key.cmd)
            
    if random.choice(chance_switch_app):
        times =random.randint(1, 5)
        print(f"switching programs {times} times")
        keyboard.press(Key.alt)
        for _ in range(1,times):
            keyboard.press(Key.tab)
            keyboard.release(Key.tab)
            time.sleep(0.25)
        keyboard.release(Key.alt)
   
    if random.choice(chance_text_to_speech):
        things_to_say=[
            'hello, are you still there?',
            'Hush now, hide, all you little ones',
            'You will bring joy to the Nowhere King, When he sees the light leaving your eyes.'
        ]
        mytext = random.choice(things_to_say)
        print(f"TTS: {mytext}")
        language = 'en'
        myobj = gTTS(text=mytext, lang=language, slow=True)

        # Saving the converted audio in a mp3 file named
        # welcome 
        myobj.save("tts.mp3")
        playsound('tts.mp3')
        
        
    if random.choice(chance_type):
        active_app = getActiveWindowAppName()
        print(f'active_app:{active_app}')
        words = []
        prepend = ''
        append = ''
        if active_app == 'teams':
            words = ['']
        if active_app == 'Roblox':
            prepend = '/'
            words = [
                '...',
                ':(',
                ':)',
                ':<',
                ':>',
                'WHAT!?',
                'why?!',
                'how do you run?',
                'how do you fly guys?!',
                'can i play with your itemzzzz',
                'free stuff plz',
                'Im new :<',
                'accepting donations'
            ]
            append = '\n'

        if len(words) > 0:
            choice = random.choice(words)
            print(f'typing {choice}')
            keyboard.type(f"{prepend}{choice}{append}")
            
    if random.choice(chance_mouse_inversion):
        start = time.time()
        how_long = random.randint(1,5)
        print(f"Inverting mouse for {how_long} seconds")
        while time.time()-start < how_long:
            m.update()
            mouse.position = m.getPos()
            time.sleep(0.01)
            
    if random.choice(chance_sound):
        print('playing random :', end='')
        with open('sound.mp3', 'wb')as f:
            res = requests.get(random.choice(sounds))
            f.write(res.content)
            print(res.url)
        playsound('sound.mp3')
        
    t = random.randint(1, 5)
    if t > 24:
        print("sneaking")
        t = random.randint(25,60)
    print(f'sleeping {t} secs')
    time.sleep(t)
```
