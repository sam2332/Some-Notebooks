```python
import zmq
import json
import time
from datetime import datetime

class RemoteZMQDict():
    def __init__(self, basket='default', host='localhost:5555'):
        self.data = {}
        self.basket = basket
        self.context = zmq.Context()
        print("Connecting to dict server…")
        self.socket = self.context.socket(zmq.REQ)
        self.socket.connect("tcp://"+host)        
        self.syncDown()
        
    def syncDown(self):
        self.socket.send_string(self.basket + ':' + '1')
        self.data.update(self.socket.recv_json())
        
    def syncUp(self):
        self.socket.send_string(self.basket + ':' + json.dumps(self.data))
        self.data.update(self.socket.recv_json()) 
        
    def delete(self,key):
        socket.send_string(self.basket + ':' + f'del {key}')
        self.data.update(self.socket.recv_json()) 
        
    def update(self,other):
        self.socket.send_string(self.basket + ':' + json.dumps(other))
        self.data.update(self.socket.recv_json())  
        
    def __getitem__(self,key):
        return self.data[key]
    
    def __setitem__(self,key,value):
        self.socket.send_string(self.basket + ':' + json.dumps({key:value}))
        self.data.update(self.socket.recv_json())  
        
    def __delitem__(self,key):
        self.socket.send_string(self.basket + ':' + f'del {key}')
        self.data.update(self.socket.recv_json()) 
        
    def __dir__(self):
        return self.data.keys()
```
