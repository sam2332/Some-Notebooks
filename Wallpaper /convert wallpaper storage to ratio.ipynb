```python
import os
from pathlib import Path
from PIL import Image
wallpaper_dir = os.path.expanduser("~/Pictures/Wallpapers3")
if not os.path.exists(wallpaper_dir):
    os.mkdir(wallpaper_dir)
```

```python
allowed=[
'.png','.jpg','.jpeg'
]
```

```python
for file in Path(os.path.expanduser("~/Pictures/Wallpapers2")).glob('**/*'):
    if file.is_file():
        if file.suffix in allowed:
            try:
                img = Image.open(file)
                height,width = img.size
                size_str =f'{width/height}'
                size_str = size_str.split('.')
                size_str = '.'.join([size_str[0],size_str[1][:2]])
                print(f"adding image to: {size_str}")
                size_folder = Path(os.path.join(wallpaper_dir,size_str))
                if not size_folder.exists():
                    size_folder.mkdir()
                out_path = os.path.join(wallpaper_dir,size_str,file.name)
                with open(out_path,'wb') as f:

                    f.write(file.read_bytes())
                    file.unlink()
            except Exception as e:
                print(e)
```
