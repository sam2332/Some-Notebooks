```python
from getpass import getpass

import keyring
import requests
import urllib3

urllib3.disable_warnings()
```

```python
cisco_hostname = ""
```

```python
username = ""
if keyring.get_password("CiscoPhoneAuth", username) is None:
    keyring.set_password("CiscoPhoneAuth", username, getpass())
```

```python
res = requests.get(
    f"https://{cisco_hostname}/vmrest/mailbox/folders/inbox/messages?mediaHint=true&rowsperpage=250&pagenumber=1",
    headers={
        "Accept": "application/json",
    },
    auth=(username, keyring.get_password("CiscoPhoneAuth", username)),
    verify=False,
).json()
```
