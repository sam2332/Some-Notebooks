```python
from pathlib import Path

p = Path("./samples")
p.mkdir(exist_ok=True)
files = list(filter(None, [f if f.is_file() else None for f in p.glob("*.*")]))
files
```

```python
file = files[1]
file
```

```python
is_raw_bytes = False
try:

    raw = file.read_text()
except UnicodeDecodeError as e:
    raw = file.read_bytes()
    is_raw_bytes = True
is_raw_bytes
```

```python
raw[:15]
```

```python
if is_raw_bytes:
    alpha_numaric = list(
        b"abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ "
    )

    hex_list = list(b"ABCDEF0123456789")
else:
    alpha_numaric = list(
        "abcdefghijklmnopqrstuvwxyz0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ "
    )
    hex_list = list("ABCDEF0123456789")
```

```python
min_length = 3
buffer = []
for i in list(raw):
    if i in hex_list:
        buffer.append(chr(i))
    else:
        if len(buffer) > min_length:
            print("".join(buffer))
        buffer = []
```
